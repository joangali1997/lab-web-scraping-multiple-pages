### 4. Display the 20 latest earthquakes info (date, time, latitude, longitude and region name) by the EMSC as a pandas dataframe: url = 'https://www.emsc-csem.org/Earthquake/'

```python
url = 'https://www.emsc-csem.org/Earthquake/'
```

```python
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')
print(soup.prettify())
```

```python
soup.select('table > tbody > tr > td.tabev6 a')[1].get_text()
```

```python
soup.select('table > tbody > tr > td')
```

```python
soup.select('table > tbody > tr > td.tabev1,td.tabev1')[0].get_text()
```

```python
soup.select('table > tbody > tr > td.tabev1')
```

```python
# Creating new lists to place: date, time, latitude, longitude and region name

date_time = []
latitude_longitude = []
region = []

num_iter = len(soup.select('table > tbody > tr'))
```

```python
# For loop to iterate through each row and append the info to each
#new list

for i in range(num_iter):
    date_time.append(soup.select('td.tabev6 a')[i].get_text())
    latitude_longitude.append(soup.select('td.tabev1')[i].get_text())
    region.append(soup.select('td.tb_region')[i].get_text())
```

```python
# Creating a pandas DataFrame with all the information requested 
earthquakes = pd.DataFrame({'Date & Time':date_time,'Region':region})
earthquakes
```
